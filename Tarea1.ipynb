### ALGUNOS CONCEPTOS DE TERMODINÁMICA

La termodinámica es el estudio científico de las relaciones entre el calor y otras formas de energía (Fegley, 2013), y sería la rama de la ciencia que se ocupa de estudiar los niveles relativos de energía y las transferencias de energía entre sistemas y entre diferentes estados de la materia.
La termodinámica es una gran ayuda para las geociencias modernas, con aplicaciones que cubren la geoquímica del interior de la Tierra, la geología marina, mineralogía, la ciencia planetaria, etc. Uno de los objetivos de la investigación geológica es comprender los procesos físicos y químicos responsables del origen y la evolución de la tierra. En términos de tiempo y espacio, la mayoría de estos procesos no son accesibles a la observación directa, pero pueden reconstruirse descifrando los marcos fisicoquímicos de formación de rocas y depósitos minerales que se formaron en respuesta a ellos. Suponiendo que el conjunto de minerales observado en una roca registran el estado de equilibrio químico de formación de la roca, la termodinámica puede ayudar a traducir los datos de composición de los minerales coexistentes a una temperatura y presión de su equilibrio determinada (Chatterjee, 1991).

#### Ley Cero
La “ley cero” también conocida como Ley del Equilibrio Térmico, dice que: 
                          $$si, A=B _y_ B=C _entonces_ A=B$$

“Si dos sistemas están en equilibrio térmico de forma independiente con un tercer sistema, deben estar también en equilibrio térmico entre sí”. Puede expresarse lógicamente del siguiente modo: si A = C y B = C, entonces A= B.

Ejemplo: Si tengo una mano fría y la otra caliente...al unirlas van a alcanzar un equilibrio donde las dos manos estén a la misma temperatura.

#### Primera ley
Esta ley dice, en terminos generales, quela cantidad de energía total del universo es constante, por lo que se le conoce como la **ley de conservación de la energía**. 
El termino energía se considera un concepto abstractom auqnue puede estar en nuestro vocabulario diario. Las deﬁniciones de energía que se encuentran comúnmente en diccionarios y libros de física establecen que la energía es la habilidad para hacer trabajo, la capacidad para hacer trabajo o el potencial para ejercer una fuerza. También se ha llamado energía a la capacidad de un sistema para provocar cambios en sí mismo o en su entorno, o la facultad inmutable de realizar un trabajo (Fegley, 2013).
El principio de conservación de la energía puede ser utilizado para definir una nueva función termodinámica E, que corresponde a la energía interna de un sistema; la cual puede adoptar muchas formas, como la energía de enlace químico, la energía eléctrica o el aumento de la energía cinética de las partículas en un gas, líquido o sólido, entre otros tipos de energía. Asimismo, la energía interna del sistema puede variar debido a condiciones externas del sistema y debe tener en cuenta las condiciones de energía potencial gravitacional de un sistema o la energía cinética del sistema cuando realiza moviento en el espacio. Po esta razón  
Por esta razóne, la primera ley se puede interpretar como un cambio infinitesimal en la energía interna (dE) de un sistema cerrado es la suma de dos términos: la ganancia (o pérdida) de una cantidad infinitesimal de calor (dq) por el sistema , y la realización de una cantidad infinitesimal de trabajo (dw) por (o sobre) el sistema (tomado de: Fegley, 2013). La ecuación de lo antes mencionado es la siguiente:

$$\Delta Energía Interna del sistema= \delta calor - \delta trabajo realizado por el sistema$$


Ejemplo: un bombillo (de los de filamente de tungteno) transforma energía eléctrica en energía calórica y energía lumínica. 



#### Segunda ley
La segunda ley de la termodinámica afirma que en cambios expontaneos el universo tiende hacia el estado de máximo desorden (Whitten, Davis, Peck, 1998),o, en mejores palabras, todo sistema aislado evoluciona a un estado de mayor información perdida. Esta ley adicona el termino de **entropia**, que corresponde, en terminos muy generales, al grado de desroden del sistema. 

La formula general de esta ley es la siguiente:

$$\Delta_{Entropia_Universo}=\Delta_{Entropia Sistema}  + \Delta_{Entropia Entorno} >0$$  $$(Proceso expontáneo)$$

Ejemplo: Un hielo dentro de un vaso de agua....el hielo tenderá a fundirse, convirtiendose en liquido.

#### Tercera ley
Esta ley define el cero absoluto de la escala de entropia. Nos dice que la entropia de una sustancia pura, perfectamente cristalina, es cero en el cero absoluto. Así, al llegar al cero absoluto (cero en unidades de grados Kelvin ), la entropía posee un valor mínimo constante y los procesos del sistema físico se detienen (Whitten, Davis, Peck, 1998). 

$0°Kelvin=-273.15°Centígrados

Ejemplo: Cuando se congela gradualmente una botella de agua hasta el cero absoluto, las molecular irán perdiendo movimiento y generaran menos calor. Cuando llegué al cero absoluto se tendra la inmovilidad total de las moléculas y no tendrán la posibilidad de generar ninguna cantidad de calor qe produzaca movimiento.
